## 问题 2

观察下方的分治算法（divide-and-conquer algorithm）的伪代码, 回答下面问题

```latex
DoSomething(A,p,r)
-----
n := r-p+1
if n=2 and A[p]>A[r] then
    swap A[p] and A[r]
else if n >= 3 then
    m = ceil(2n/3)
    DoSomething(A,p,p+m-1)
    DoSomething(A,r-m+1,r)
    DoSomething(A,p,p+m-1)  
    
---
first call: DoSomething(A,1,n)
```

note：$ceil(2n/3)=\left\lceil {2n/3} \right\rceil$；$:=$ 表示赋值, 等价于 $\to$；A是一个包含n的整数元素的数组, 

1. 写出该算法时间复杂度的递归公式, 并求解其对应的渐进表示
2. 描述一下该算法的功能, 并判断是否是最高效的解决方案
3. \* 使用python编程实现上述算法或其对应的更高效版本的算法

该算法的功能是对数组 A[p...r] 进行升序排序。
算法分析：
设当前处理的数组长度为 n = r - p + 1。
当 n = 2 时，如果 A[p] > A[r]，就交换这两个元素，因此长度为 2 的数组可以被排好序。
当 n >= 3 时，算法令 m = ceil(2n/3)，然后递归执行三次：
1. 对前 2/3 部分排序；
2. 对后 2/3 部分排序；
3. 再对前 2/3 部分排序。
因此，每一层递归会产生 3 个子问题，每个子问题的规模大约是 2n/3，每一层除了递归以外只做常数次操作。
所以时间复杂度递归式为：
T(n) = 3T(2n/3) + O(1)
递归树的深度大约是 log_{3/2} n。
叶子节点数量大约是：
3^(log_{3/2} n) = n^(log_{3/2} 3)
因此：
T(n) = Θ(n^(log_{3/2} 3))
因为 log_{3/2} 3 约等于 2.71，所以该算法的时间复杂度约为：
T(n) = Θ(n^2.71)
空间复杂度：
由于算法是递归实现的，递归深度为 O(log n)，所以空间复杂度为：
O(log n)
该算法不是最高效的排序算法。更高效的排序算法如归并排序、快速排序、堆排序等，时间复杂度一般是 O(n log n)。Python 内置的 sort() 也比该算法高效很多。
最终答案：
递归式：T(n) = 3T(2n/3) + O(1)

空间复杂度：O(log n)

算法功能：对数组进行升序排序

时间复杂度：Θ(n^2.71)


idea:

````python
import math

def do_something(A, p, r):
    """
    对数组 A[p...r] 进行排序。
    这里的 p 和 r 使用 Python 的下标，从 0 开始。
    """
    n = r - p + 1
    
    if n == 2 and A[p] > A[r]:
        A[p], A[r] = A[r], A[p]
    
    elif n >= 3:
        m = math.ceil(2 * n / 3)
        
        do_something(A, p, p + m - 1)
        do_something(A, r - m + 1, r)
        do_something(A, p, p + m - 1)


# 测试
A = [5, 2, 4, 1, 3]
print("排序前:", A)

do_something(A, 0, len(A) - 1)

print("排序后:", A)

# 时间复杂度：T(n) = 3T(2n/3) + O(1)
# 渐近复杂度：Θ(n^(log_{3/2}3))，约为 Θ(n^2.71)
# 空间复杂度：O(log n)
#本题部分思路参考了 ChatGPT 的辅助解释，最终答案由本人理解后整理完成。
````

## 问题 3

给定一个包含n个元素的数组 `profits` , 它的第 `i` 个元素 `profits[i]` 表示一支股票第 `i` 天的**收益**（正数表示涨, 负数表示跌）。你只能选择 **某一天** 买入这只股票, 并选择在 **未来的某一个不同的日子** 卖出该股票。

1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。

2. \* 设计一个时间复杂度为 $O(n)$的算法实现该算法

e.g. :
---
profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9

问题 3 解答

这道题可以转化为“最大连续子数组和”问题。
数组 profits 中的每个元素表示某一天的收益。  
如果选择第 i 天买入，第 j 天卖出，那么总收益就是 profits[i] 到 profits[j] 这一段连续区间的和。
因此，问题变成：
在数组 profits 中找到一个连续区间，使得这个区间的元素和最大。
可以使用 Kadane 算法。
算法思想：
从左到右扫描数组，维护两个变量：
1. current_sum：表示以当前这一天结尾时，能够获得的最大连续收益；
2. max_sum：表示到目前为止找到的最大收益。
对于当前元素 profits[i]：
如果前面的 current_sum 小于 0，说明前面的收益会拖累当前收益，所以应该从当前这一天重新开始买入。
如果前面的 current_sum 大于等于 0，说明前面的收益对当前结果有帮助，就继续累加。
同时记录最大收益对应的开始日期和结束日期。
时间复杂度：
只需要遍历数组一次，所以时间复杂度是 O(n)。
空间复杂度：
只使用了几个变量，所以空间复杂度是 O(1)。

idea:

````python
def max_profit(profits):
    """
    使用 Kadane 算法求最大连续子数组和。
    返回最大收益、买入日期、卖出日期。
    日期按照题目习惯，从 1 开始计数。
    """
    
    # 当前连续区间的收益
    current_sum = profits[0]
    
    # 当前找到的最大收益
    max_sum = profits[0]
    
    # 当前连续区间的开始位置
    current_start = 0
    
    # 最大收益区间的开始和结束位置
    best_start = 0
    best_end = 0
    
    for i in range(1, len(profits)):
        
        # 如果前面的收益为负数，那么不如从当前这一天重新开始
        if current_sum < 0:
            current_sum = profits[i]
            current_start = i
        else:
            current_sum += profits[i]
        
        # 更新最大收益
        if current_sum > max_sum:
            max_sum = current_sum
            best_start = current_start
            best_end = i
    
    # 日期从 1 开始，所以返回时下标加 1
    return max_sum, best_start + 1, best_end + 1


# 测试
profits = [3, 2, 1, -7, 5, 2, -1, 3, -1]

max_sum, buy_day, sell_day = max_profit(profits)

print("最大收益：", max_sum)
print("买入日期：第", buy_day, "天")
print("卖出日期：第", sell_day, "天")

# 时间复杂度：O(n)
# 空间复杂度：O(1)
#本题部分思路参考了 ChatGPT 的辅助解释，最终答案由本人理解后整理完成。
````

## 问题 4

给定一个包含不同整数元素的数组 $ A[1..n]$ ,并且满足条件：$A[1]>A[2]$ 并且 $A[n-1]<A[n]$; 规定：如果一个元素比它两边的邻居元素都小, 即：$A[x]<A[x-1], A[x]<A[x+1]$ , 称这个元素A[x]为“局部最小”。通过遍历一次数组, 我们可以很容易在 $O(n)$的时间复杂度下找到一个局部最小值, 


1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现
2. \* 设计算法找出所有局部最小值, 分析算法设计思路, 并使用python编程实现

e.g.:
---
A=[9, 3, 7, 2, 1, 4, 5 ] 时,  局部最小元素为 3, 1

问题 4 解答
题目给定一个数组 A[1...n]，并且满足：
A[1] > A[2]
A[n-1] < A[n]
局部最小值的定义是：如果某个元素 A[x] 同时小于它左右两边的元素，即：
A[x] < A[x-1] 且 A[x] < A[x+1]
那么 A[x] 就是一个局部最小值。
### 第 1 问：O(log n) 找到一个局部最小值
可以使用二分查找。
因为题目给出 A[1] > A[2]，说明数组开头是下降趋势。
又因为 A[n-1] < A[n]，说明数组结尾是上升趋势。
从下降到上升的过程中，一定会出现一个局部最小值。
设 mid 是当前区间的中点。
如果：
A[mid] < A[mid-1] 且 A[mid] < A[mid+1]
那么 A[mid] 就是局部最小值，直接返回。
如果：
A[mid] > A[mid-1]
说明 mid 左边有更小的元素，因此左半部分一定存在局部最小值。
如果：
A[mid] > A[mid+1]
说明 mid 右边有更小的元素，因此右半部分一定存在局部最小值。
每次查找都可以排除一半区间，所以时间复杂度是 O(log n)。
空间复杂度是 O(1)。
### 第 2 问：找出所有局部最小值
如果要找出所有局部最小值，就不能只用二分查找。
因为二分查找只能找到其中一个局部最小值。
要找出所有局部最小值，可以直接遍历数组中间的每一个元素。
对于每个位置 i，判断它是否满足：
A[i] < A[i-1] 且 A[i] < A[i+1]
如果满足，就把 A[i] 加入答案。
因为每个元素只检查一次，所以时间复杂度是 O(n)。
空间复杂度是 O(k)，其中 k 是局部最小值的个数。

idea:

````python
def find_one_local_min(A):
    """
    第 1 问：
    使用二分查找，在 O(log n) 时间内找到一个局部最小值。
    返回局部最小值的数值。
    
    注意：
    题目中的数组下标从 1 开始，
    Python 中的数组下标从 0 开始。
    """
    n = len(A)

    left = 1
    right = n - 2

    while left <= right:
        mid = (left + right) // 2

        # 如果 A[mid] 比左右邻居都小，那么它就是局部最小值
        if A[mid] < A[mid - 1] and A[mid] < A[mid + 1]:
            return A[mid]

        # 如果左边元素更小，说明左半部分一定存在局部最小值
        elif A[mid] > A[mid - 1]:
            right = mid - 1

        # 如果右边元素更小，说明右半部分一定存在局部最小值
        else:
            left = mid + 1

    return None


def find_all_local_min(A):
    """
    第 2 问：
    遍历数组，找出所有局部最小值。
    返回所有局部最小值的数值。
    """
    n = len(A)
    result = []

    for i in range(1, n - 1):
        if A[i] < A[i - 1] and A[i] < A[i + 1]:
            result.append(A[i])

    return result


# 测试
A = [9, 3, 7, 2, 1, 4, 5]

one_local_min = find_one_local_min(A)
all_local_min = find_all_local_min(A)

print("数组：", A)
print("找到的一个局部最小值：", one_local_min)
print("所有局部最小值：", all_local_min)

# 第 1 问时间复杂度：O(log n)
# 第 1 问空间复杂度：O(1)

# 第 2 问时间复杂度：O(n)
# 第 2 问空间复杂度：O(k)，k 是局部最小值的个数
#本题部分思路参考了 ChatGPT 的辅助解释，最终答案由本人理解后整理完成。
````

## 问题 5

给定包含n个不同数字的一组数, 寻找一种基于比较的算法在这组数中找到k个最小的数字, 并按顺序输出它们。

1. 将n个数先进行排序, 然后按顺序输出最小的k个数。要求：选择合适的排序算法实现上述操作, 计算算法时间复杂度, 并使用python编程实现。
2. 建立一个包含这n个数的堆（heap）, 并且调用 k 次Extract-min 按顺序输出最小的k个数。使用往空堆中不断插入元素的方法建立堆, 分析这种方法建堆的时间复杂度, 并使用python编程实现
3. \* 假设数组中包含的数据总数目超过了计算机的存储能力, 请设计一个算法, 找到这堆数据的前k小的数值, 计算时间复杂度, 并使用python实现该算法, 假设计算机一定能存储k个数据。

e.g.：
---
数组arr=[5,4,3,2,6,1,88,33,22,107] 的前3个最小数据为：1, 2, 3

### 问题 5 解答

本题要求在一个包含 n 个不同数字的数组中，找到最小的 k 个数字，并按照升序输出。
### 第 1 问：排序后输出前 k 个数
最直接的方法是先将整个数组排序，然后输出排序后的前 k 个元素。
算法步骤：
1. 对数组 arr 进行升序排序；
2. 取出排序后数组的前 k 个元素；
3. 输出这 k 个元素。
时间复杂度：
排序需要 O(n log n)，取前 k 个元素需要 O(k)。
所以总时间复杂度为：
O(n log n)
空间复杂度：
如果使用 Python 的 sorted() 函数，会产生一个新的排序数组，因此空间复杂度为 O(n)。
### 第 2 问：建立最小堆，然后连续 extract-min
可以使用最小堆。
算法步骤：
1. 建立一个空的最小堆；
2. 将数组中的 n 个元素依次插入最小堆；
3. 连续执行 k 次 extract-min 操作，每次取出当前堆中的最小值；
4. 取出的 k 个数就是数组中最小的 k 个数，并且天然是升序的。
时间复杂度：
向堆中插入 n 个元素，每次插入的代价是 O(log n)，所以建堆时间为：
O(n log n)
连续取出 k 次最小值，每次 extract-min 的代价是 O(log n)，所以取出 k 个最小值的时间为：
O(k log n)
因此总时间复杂度为：
O((n + k) log n)
空间复杂度：
堆中最多存储 n 个元素，所以空间复杂度是 O(n)。
### 第 3 问：数据量太大，内存只能存储 k 个数据
如果数据总量 E 太大，不能把所有数据一次性放入内存，但是计算机能够存储 k 个数据，那么可以使用大小为 k 的最大堆。
算法思想：
维护一个大小为 k 的最大堆，堆中始终保存当前已经扫描过的数据中最小的 k 个数。
算法步骤：
1. 先读取前 k 个元素，建立一个大小为 k 的最大堆；
2. 继续依次读取后面的元素 x；
3. 如果 x 大于等于堆顶元素，说明 x 不可能属于最小的 k 个数，直接跳过；
4. 如果 x 小于堆顶元素，说明 x 应该进入答案，于是删除堆顶元素，并把 x 插入堆中；
5. 扫描完所有数据后，堆中保存的就是最小的 k 个数；
6. 最后将堆中的 k 个数排序后输出。
这里使用最大堆的原因是：
最大堆的堆顶是当前 k 个候选数中最大的那个。  
如果新来的元素比堆顶还大，那么它一定不可能进入前 k 小。  
如果新来的元素比堆顶小，就可以替换堆顶。
时间复杂度：
总共有 E 个数据，每个数据最多进行一次堆调整，堆的大小始终是 k，所以时间复杂度为：
O(E log k)
最后对 k 个元素排序需要 O(k log k)。
因此总时间复杂度为：
O(E log k + k log k)
空间复杂度：
内存中只保存 k 个元素，所以空间复杂度为：
O(k)
### 最终答案：
第 1 问：
排序法，时间复杂度 O(n log n)，空间复杂度 O(n)。
第 2 问：
最小堆，插入 n 个元素，再 extract-min k 次。
时间复杂度 O((n+k)log n)，空间复杂度 O(n)。
第 3 问：
大数据场景，维护大小为 k 的最大堆。
时间复杂度 O(E log k + k log k)，空间复杂度 O(k)。

idea:

````python
import heapq


def method1_sort(arr, k):
    """
    方法一：
    先排序，再取前 k 个最小值。
    时间复杂度：O(n log n)
    空间复杂度：O(n)
    """
    sorted_arr = sorted(arr)
    return sorted_arr[:k]


def method2_min_heap(arr, k):
    """
    方法二：
    建立最小堆，然后连续取出 k 次最小值。
    时间复杂度：O((n + k) log n)
    空间复杂度：O(n)
    """
    heap = []
    
    # 依次插入元素建立最小堆
    for x in arr:
        heapq.heappush(heap, x)
    
    result = []
    
    # 连续 extract-min k 次
    for _ in range(k):
        result.append(heapq.heappop(heap))
    
    return result


def method3_big_data(arr, k):
    """
    方法三：
    当数据量太大，内存只能存储 k 个数据时，
    使用大小为 k 的最大堆维护当前最小的 k 个数。
    
    Python 默认只有最小堆，
    所以这里用负数来模拟最大堆。
    
    时间复杂度：O(E log k + k log k)
    空间复杂度：O(k)
    """
    max_heap = []
    
    for x in arr:
        if len(max_heap) < k:
            # 插入 -x，用最小堆模拟最大堆
            heapq.heappush(max_heap, -x)
        else:
            # 当前堆顶 -max_heap[0] 是这 k 个数中最大的那个
            current_max = -max_heap[0]
            
            if x < current_max:
                # 删除当前最大的数，插入更小的 x
                heapq.heapreplace(max_heap, -x)
    
    # max_heap 中存的是负数，需要转回正数
    result = [-x for x in max_heap]
    
    # 最后按升序输出
    result.sort()
    
    return result


# 测试
arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
k = 3

print("原数组：", arr)
print("k =", k)

print("方法一，排序后取前 k 个：", method1_sort(arr, k))
print("方法二，最小堆连续 extract-min：", method2_min_heap(arr, k))
print("方法三，大数据情况下大小为 k 的最大堆：", method3_big_data(arr, k))

# 方法一时间复杂度：O(n log n)
# 方法二时间复杂度：O((n + k) log n)
# 方法三时间复杂度：O(E log k + k log k)
#本题部分思路参考了 ChatGPT 的辅助解释，最终答案由本人理解后整理完成。
````

## 问题 6

> 星球碰撞问题：现有n个星球，在同一条直线上运行，如数组A所示，元素的绝对值表示星球的质量，负数表示星球自右向左运动，正数表示星球自左向右运动，当两个星球相撞的时候，质量小的会消失，大的保持不变，**质量相同的两个星球碰撞后自右向左运动的星球消失，自左向右的星球保持不变**，假设所有星球的速度大小相同。
>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]

### 问题 6 解答
本题是星球碰撞问题。
数组 A 中每个元素表示一个星球：
- 绝对值表示星球质量；
- 正数表示星球从左向右运动；
- 负数表示星球从右向左运动。
只有一种情况会发生碰撞：
左边的星球向右运动，右边的星球向左运动。
也就是说：
前面的星球是正数，后来的星球是负数。
例如：
8 和 -5 会碰撞。
但是：
- -3 和 -6 不会碰撞，因为它们都向左运动；
- -3 和 5 不会碰撞，因为它们方向相反但正在远离；
- 3 和 5 不会碰撞，因为它们都向右运动。
### 算法思想
使用一个栈 stack 保存当前还没有被撞毁的星球。
从左到右遍历数组 A。
对于当前星球 x：
1. 如果 x 是正数，直接入栈；
2. 如果 x 是负数，并且栈顶是正数，说明会发生碰撞；
3. 比较栈顶星球和当前星球的质量大小：
   - 如果栈顶星球质量小，那么栈顶星球消失，继续比较；
   - 如果栈顶星球质量大，那么当前星球消失；
   - 如果质量相同，根据题目规则，向右运动的星球保留，向左运动的星球消失，也就是当前负数星球消失；
4. 如果当前星球没有被撞毁，就入栈。
最后栈中剩下的星球就是最终结果。
### 时间复杂度分析
每个星球最多入栈一次、出栈一次。
所以时间复杂度是：
O(n)
### 空间复杂度分析
最坏情况下所有星球都不会碰撞，都保存在栈中。
所以空间复杂度是：O(n)

idea:

````python
def asteroid_collision(A):
    """
    模拟星球碰撞过程。
    
    正数：从左向右运动
    负数：从右向左运动
    
    当正数星球和负数星球相遇时：
    质量小的消失；
    质量大的保留；
    质量相同的时候，负数星球消失，正数星球保留。
    """
    stack = []
    
    for x in A:
        alive = True
        
        # 只有当当前星球 x 是负数，且栈顶星球是正数时，才会发生碰撞
        while alive and x < 0 and stack and stack[-1] > 0:
            
            # 栈顶正数星球质量小，会消失
            if stack[-1] < -x:
                stack.pop()
            
            # 两个星球质量相同
            # 按照题目规则：从右向左运动的星球消失
            # 当前 x 是负数，所以当前星球消失
            elif stack[-1] == -x:
                alive = False
            
            # 栈顶正数星球质量更大，当前负数星球消失
            else:
                alive = False
        
        # 如果当前星球没有被撞毁，则入栈
        if alive:
            stack.append(x)
    
    return stack


# 测试 1
A1 = [-3, -6, 2, 8, 5, -8, 9, -2, 1]
print("输入：", A1)
print("输出：", asteroid_collision(A1))

# 测试 2
A2 = [23, -8, 9, -3, -7, 9, -23, 22]
print("输入：", A2)
print("输出：", asteroid_collision(A2))

# 时间复杂度：O(n)
# 空间复杂度：O(n)
#本题部分思路参考了 ChatGPT 的辅助解释，最终答案由本人理解后整理完成。
````

## 问题 7

> 0-1背包问题：现在有4块大理石，每个大理石对应的重量和价值使用一个元组表示, 即（重量，价值），4块大理石的重量和价值为：[(5,10), (4,40), (6,30), (3, 50)], 假设你有一辆最大承重为9的小推车，请问使用这个小推车装大理石的最大价值为多少

请设计一个算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

idea:

### 问题 7 解答
本题是一个 0-1 背包问题。
每块大理石都有重量和价值，每块大理石只能选择一次，要么放入小推车，要么不放入小推车。
题目要求在小推车最大承重为 9 的情况下，选择若干块大理石，使得总重量不超过 9，并且总价值最大。
### 算法思想
使用动态规划。
设：
dp[i][w] 表示只考虑前 i 个物品，并且背包容量为 w 时，可以获得的最大价值。
对于第 i 个物品，它有两种选择：
1. 不选第 i 个物品  
   此时最大价值为：
   dp[i-1][w]
2. 选第 i 个物品  
   前提是当前容量 w 能够装下它，也就是 w >= weight[i]。  
   此时最大价值为：
   dp[i-1][w-weight[i]] + value[i]
因此状态转移方程为：
如果 w < weight[i]：
dp[i][w] = dp[i-1][w]
如果 w >= weight[i]：
dp[i][w] = max(dp[i-1][w], dp[i-1][w-weight[i]] + value[i])
### 时间复杂度分析
一共有 n 个物品，背包容量为 W。
动态规划需要枚举每个物品和每个容量。
所以时间复杂度是：
O(nW)
### 空间复杂度分析
使用二维 dp 数组，大小为：
(n+1) × (W+1)
所以空间复杂度是：
O(nW)
### 本题结果
物品为：
(5,10), (4,40), (6,30), (3,50)
背包容量为 9。
最优选择是：
重量为 4、价值为 40 的大理石
和
重量为 3、价值为 50 的大理石
总重量为：
4 + 3 = 7
总价值为：
40 + 50 = 90
所以最大价值是 90。

算法：动态规划
状态定义：
dp[i][w] 表示考虑前 i 个物品，容量为 w 时能获得的最大价值。
状态转移：
dp[i][w] = max(dp[i-1][w], dp[i-1][w-weight[i]] + value[i])
时间复杂度：
O(nW)
空间复杂度：
O(nW)
本题最大价值：
90

````python
def knapsack_01(items, capacity):
    """
    0-1 背包问题
    
    items 中的每个元素是一个二元组：
    (weight, value)
    
    capacity 表示背包最大容量。
    """
    n = len(items)
    
    # dp[i][w] 表示考虑前 i 个物品，容量为 w 时的最大价值
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]
    
    for i in range(1, n + 1):
        weight, value = items[i - 1]
        
        for w in range(capacity + 1):
            # 情况 1：不选择第 i 个物品
            dp[i][w] = dp[i - 1][w]
            
            # 情况 2：选择第 i 个物品
            if w >= weight:
                dp[i][w] = max(dp[i][w], dp[i - 1][w - weight] + value)
    
    return dp[n][capacity]


# 测试
items = [(5, 10), (4, 40), (6, 30), (3, 50)]
capacity = 9

answer = knapsack_01(items, capacity)

print("物品列表：", items)
print("背包容量：", capacity)
print("最大价值：", answer)

# 时间复杂度：O(nW)，其中 n 是物品数量，W 是背包容量
# 空间复杂度：O(nW)
#本题部分思路参考了 ChatGPT 的辅助解释，最终答案由本人理解后整理完成。
````

## 问题 8

给你二叉搜索树的根节点 root ，该树中的恰好两个节点的值被错误地交换。请在不改变其结构的情况下，恢复这棵树 。设计算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

e.g.:
---
1.

<div style="text-align:center;width:200px;"> <img alt="" src="./fig/hw3q9-1.jpg"> </div>
输入：root = [1,3,null,null,2]， 输出为：[3,1,null,null,2]， 

解释：3 不能是 1 的左孩子，因为 3 > 1 。交换 1 和 3 使二叉搜索树有效。

2.

 <div style="text-align:center;width:200px;"> <img alt="" src="./fig/hw3q9-1.jpg"> </div>
输入： root=[3,1,4,null,null,2]， 输出为：[2,1,4,null,null,3]

解释：2 不能在 3 的右子树中，因为 2 < 3 。交换 2 和 3 使二叉搜索树有效。

idea:

### 问题 8 解答
本题要求恢复一棵二叉搜索树。
二叉搜索树有一个重要性质：
对二叉搜索树进行中序遍历，得到的结果应该是严格递增的。
如果有两个节点的值被错误交换，那么中序遍历序列中就会出现逆序。
例如：
正常 BST 的中序遍历应该是：
1, 2, 3, 4
如果 2 和 4 被交换，可能变成：
1, 4, 3, 2
此时会出现逆序：
4 > 3
3 > 2
我们只需要在中序遍历过程中找到这两个错误的节点，然后交换它们的值即可。
### 算法思想
使用中序遍历。
在遍历过程中，用 prev 记录当前节点的前一个节点。
如果发现：
prev.val > current.val
说明这两个节点的顺序不符合二叉搜索树的中序递增性质。
需要记录两个错误节点：
1. 第一次发现逆序时，prev 是第一个错误节点；
2. 每次发现逆序时，current 是第二个错误节点。
最后交换这两个节点的值。
### 时间复杂度分析
每个节点只访问一次。
所以时间复杂度是：
O(n)
其中 n 是二叉树中的节点个数。
### 空间复杂度分析
使用递归中序遍历，递归栈的深度与树的高度有关。
所以空间复杂度是：
O(h)
其中 h 是二叉树的高度。
如果树退化成链表，最坏情况下空间复杂度是 O(n)。

算法：中序遍历
关键性质：
二叉搜索树的中序遍历结果应该是递增的。
做法：
找到中序遍历中出现逆序的两个节点，然后交换它们的值。
时间复杂度：
O(n)
空间复杂度：
O(h)

````python
from collections import deque


class TreeNode:
    """
    二叉树节点定义
    """
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def recover_tree(root):
    """
    恢复被错误交换两个节点值的二叉搜索树。
    不改变树的结构，只交换两个节点的值。
    """
    first = None
    second = None
    prev = None

    def inorder(node):
        nonlocal first, second, prev

        if node is None:
            return

        # 中序遍历左子树
        inorder(node.left)

        # 如果前一个节点的值大于当前节点的值，说明出现逆序
        if prev is not None and prev.val > node.val:
            if first is None:
                first = prev
            second = node

        # 更新 prev
        prev = node

        # 中序遍历右子树
        inorder(node.right)

    inorder(root)

    # 交换两个错误节点的值
    if first is not None and second is not None:
        first.val, second.val = second.val, first.val


def build_tree(values):
    """
    根据层序列表构造二叉树。
    例如：[1, 3, None, None, 2]
    """
    if not values or values[0] is None:
        return None

    root = TreeNode(values[0])
    queue = deque([root])
    i = 1

    while queue and i < len(values):
        node = queue.popleft()

        if i < len(values) and values[i] is not None:
            node.left = TreeNode(values[i])
            queue.append(node.left)
        i += 1

        if i < len(values) and values[i] is not None:
            node.right = TreeNode(values[i])
            queue.append(node.right)
        i += 1

    return root


def tree_to_list(root):
    """
    将二叉树按层序遍历转换成列表，方便输出查看。
    """
    if root is None:
        return []

    result = []
    queue = deque([root])

    while queue:
        node = queue.popleft()

        if node is None:
            result.append(None)
        else:
            result.append(node.val)
            queue.append(node.left)
            queue.append(node.right)

    # 去掉末尾多余的 None
    while result and result[-1] is None:
        result.pop()

    return result


# 测试 1
root1 = build_tree([1, 3, None, None, 2])
print("恢复前：", tree_to_list(root1))

recover_tree(root1)

print("恢复后：", tree_to_list(root1))


# 测试 2
root2 = build_tree([3, 1, 4, None, None, 2])
print("恢复前：", tree_to_list(root2))

recover_tree(root2)

print("恢复后：", tree_to_list(root2))


# 时间复杂度：O(n)
# 空间复杂度：O(h)
#本题部分思路参考了 ChatGPT 的辅助解释，最终答案由本人理解后整理完成。
````

````python
from collections import deque


class TreeNode:
    """
    二叉树节点定义
    """
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def sum_numbers(root):
    """
    计算所有根节点到叶子节点路径所表示数字之和。
    """

    def dfs(node, current):
        if node is None:
            return 0

        # 更新当前路径形成的数字
        current = current * 10 + node.val

        # 如果是叶子节点，返回当前路径形成的数字
        if node.left is None and node.right is None:
            return current

        # 继续计算左子树和右子树
        left_sum = dfs(node.left, current)
        right_sum = dfs(node.right, current)

        return left_sum + right_sum

    return dfs(root, 0)


def build_tree(values):
    """
    根据层序遍历列表构造二叉树。
    例如：[1, 2, 3]
    """
    if not values or values[0] is None:
        return None

    root = TreeNode(values[0])
    queue = deque([root])
    i = 1

    while queue and i < len(values):
        node = queue.popleft()

        if i < len(values) and values[i] is not None:
            node.left = TreeNode(values[i])
            queue.append(node.left)
        i += 1

        if i < len(values) and values[i] is not None:
            node.right = TreeNode(values[i])
            queue.append(node.right)
        i += 1

    return root


# 测试 1
root1 = build_tree([1, 2, 3])
print("输入：root = [1, 2, 3]")
print("输出：", sum_numbers(root1))


# 测试 2
root2 = build_tree([4, 9, 0, 5, 1])
print("输入：root = [4, 9, 0, 5, 1]")
print("输出：", sum_numbers(root2))


# 时间复杂度：O(n)
# 空间复杂度：O(h)
#本题部分思路参考了 ChatGPT 的辅助解释，最终答案由本人理解后整理完成。
````
